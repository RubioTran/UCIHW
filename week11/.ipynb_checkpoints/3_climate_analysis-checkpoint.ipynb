```python
con = sqlite3.connect("hawaii.sqlite")
prcp_df = pd.read_sql_query("SELECT date, prcp FROM measurements WHERE date >= '2016-01-01' AND date <= '2016-12-31'", con)
print(prcp_df.head(2))
con.close()
```

```python
prcp_df.set_index('date')
print(prcp_df.head(2))
print(prcp_df.tail(2))
```

```python
#station analysis
```

```python
con = sqlite3.connect("hawaii.sqlite")
stations_df = pd.read_sql_query("SELECT COUNT(name) AS 'Total Number of Stations' FROM stations", con)
print(stations_df.head())
con.close()
```

```python
con = sqlite3.connect("hawaii.sqlite")
stations_df = pd.read_sql_query(\
"SELECT measurements.station, stations.name, COUNT(measurements.date) AS 'Obs Counts' \
FROM measurements INNER JOIN stations \
WHERE measurements.station = stations.station GROUP BY stations.name ORDER BY COUNT(measurements.date) DESC"\
, con)
print(stations_df.head(10))
con.close()
```

```python
print(str(stations_df.iloc[0,1]) + "has the most observations at " + str(stations_df.iloc[0,2]))
```

```python
con = sqlite3.connect("hawaii.sqlite")
tobs_df = pd.read_sql_query("SELECT station, COUNT(tobs) as 'Count of TOBS' FROM measurements \
WHERE date >= '2016-01-01' AND date <= '2016-12-31'GROUP BY station ORDER BY COUNT(tobs) DESC", con)
print(tobs_df)
con.close()
```

```python
tobs_df.hist(bins=12)
plt.show()
```
